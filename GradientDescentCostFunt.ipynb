```python
# import neccessary libraries

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def gradient_descent(x,y,learning_rate,iterate):
    m_curr = b_curr = 0
    iterations = iterate
    n= len(x)
    learning_rate = learning_rate
  
    for i in range(iterations):
        y_pred = m_curr*x + b_curr
        md = -(2/n)*sum(x*(y-y_pred))
        bd = -(2/n)*sum(y-y_pred)
        
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        cost = (1/n) * sum([ val**2 for val in (y-y_pred)])
        
      
        
        print(f" m : {m_curr}  , md {md}  bd {bd} and b: {b_curr} ,cost : {cost} , iteration : {i}")
        
        

def gradient_plot(x,y,learning_rate,iterate):
    m_curr = b_curr = 0
    iterations = iterate
    n= len(x)
    learning_rate = learning_rate
    
    grad_dict ={}
    

    
    for i in range(iterations):
        y_pred = m_curr*x + b_curr
        md = -(2/n)*sum(x*(y-y_pred))
        bd = -(2/n)*sum(y-y_pred)
        
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        cost = (1/n) * sum([ val**2 for val in (y-y_pred)])
        
        
        plt.plot(x,y_pred,color = 'green')
        
    plt.xlabel("independent value")
    plt.ylabel("dependent value")
    
    
    


```

```python
# calculate mean square error and plot with m and b
def errors(x_values, y_values, m, b):
    y_line = (b + m*x_values)
    return (y_values - y_line)

def squared_errors(x_values, y_values, m, b):
    return np.round(errors(x_values, y_values, m, b)**2, 2)

def residual_sum_squares(x_values, y_values, m, b):
    return round(sum(squared_errors(x_values, y_values, m, b)), 2)

def mean_square_error(x_values, y_values, m, b):
    n = len(x_values)
    return round((residual_sum_squares(x_values, y_values, m, b))/n, 2)


# taking random variable for x and y

x = np.random.rand(30,1).reshape(30)
y_randterm = np.random.normal(0,3,30)
y = 3+ 50* x + y_randterm


table = np.zeros((20,2))
for idx, val in enumerate(range(40, 60)):
    table[idx,0] = val
    table[idx,1] = mean_square_error(x, y, val, 2.9)
    
table_b = np.zeros((12,2))
for idx, b_val in enumerate(range(-2, 10)):
    table_b[idx,0] = b_val
    table_b[idx,1] = mean_square_error(x, y, 48, b_val) 
    
    
    
fig , ax = plt.subplots(1,2,figsize=(18,7))    
    
#plt.figure(figsize=(10,7))
ax[0].plot(table[:,0], table[:,1], '-')
ax[0].set_xlabel("m-values", fontsize=14)
ax[0].set_ylabel("MSE", fontsize=14)
ax[0].set_title("MSE with changes to slope", fontsize=16);


ax[1].plot(table_b[:,0], table_b[:,1], '-')
ax[1].set_xlabel("b-values", fontsize=14)
ax[1].set_ylabel("MSE", fontsize=14)
ax[1].set_title("MSE with changes to intercept", fontsize=16);
```
